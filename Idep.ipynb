```python
#Excluir Colunas  de 2023 
df_2023 = df_2023.drop(["IDEP_INICIAL.1", "IDEP_FINAL.1"], axis=1)
```

```python
#Criando Analises e colunas para base 2023
display(df_2023)
```

```python
#Media e Mediana  do ano  2023
media_inicial = df_2023["IDEP_INICIAL"].mean()
media_final = df_2023["IDEP_FINAL"].mean()

print(f"Média do IDEP_INICIAL de 2023: {media_inicial:.2f}")
print(f"Média do IDEP_FINAL de 2023: {media_final:.2f}")

#Mediana
mediana_inicial = df_2023["IDEP_INICIAL"].median()
mediana_final = df_2023["IDEP_FINAL"].median()

print(f"Mediana do IDEP_INICIAL de 2023: {mediana_inicial:.2f}")
print(f"Mediana do IDEP_FINAL de 2023: {mediana_final:.2f}")
```

```python
# Calculando a diferença entre o IDEP INCIAL e IDEP FINAL de 2023
tabela_2023 = df_2023.groupby("Faixa").agg({
    "IDEP_INICIAL": ["mean", "median", "std"],
    "IDEP_FINAL": ["mean", "median", "std"],
    
}).round(2)

display(tabela_2023)

```

```python
#Media e Mediana  da minha base 2024 

media_inicial_2024 = df_2024["IDEP_INICIAL"].mean()
media_final_2024= df_2024["IDEP_FINAL"].mean()
print(f"Média do IDEP_INICIAL de 2024: {media_inicial_2024:.2f}")
print(f"Média do IDEP_FINA de 2024: {media_final_2024:.2f}")




```

```python
#Media , Mediana e desvio padrao de 2024 

tabela_2024 = df_2024.groupby("Faixa").agg({
    "IDEP_INICIAL": ["mean", "median", "std"],
    "IDEP_FINAL": ["mean", "median", "std"],
    
}).round(2)

display(tabela_2024)
```

```python
df_2023["Faixa_NSE"] = pd.cut(
    df_2023["NSE"],
    bins=[0, 40, 50, 60, 70, 100],
    labels=["Muito baixo", "Baixo", "Médio", "Médio alto", "Alto"]
)
```

```python
#Criando as Colunas para dentro da tabela 2024
df_2024["evolucao_IDEP"] = (df_2024["IDEP_FINAL"] - df_2024["IDEP_INICIAL"]).round(2)

condicoes = [
    df_2024["evolucao_IDEP"] >= 0.5,
    (df_2024["evolucao_IDEP"] >= 0.2) & (df_2024["evolucao_IDEP"] < 0.5),
    (df_2024["evolucao_IDEP"] > -0.2) & (df_2024["evolucao_IDEP"] < 0.2),
    (df_2024["evolucao_IDEP"] >= -0.5) & (df_2024["evolucao_IDEP"] <= -0.2),
    df_2024["evolucao_IDEP"] < -0.5
]

categorias = [
    "Avanço significativo",
    "Avanço moderado",
    "Estabilidade",
    "Declínio moderado",
    "Declínio significativo"
]

df_2024["categoria_evolucao"] = np.select(condicoes, categorias, default="Sem classificação")

df_2024["crescimento_percentual"] = df_2024["crescimento_percentual"].round(2)
```

```python

#Crianndo Novas colunas de Metrica de
df_2023["evolucao_IDEP"] = df_2023["IDEP_FINAL"] - df_2023["IDEP_INICIAL"]

condicoes = [
    df_2023["evolucao_IDEP"] >= 0.5,
    (df_2023["evolucao_IDEP"] >= 0.2) & (df_2023["evolucao_IDEP"] < 0.5),
    (df_2023["evolucao_IDEP"] > -0.2) & (df_2023["evolucao_IDEP"] < 0.2),
    (df_2023["evolucao_IDEP"] >= -0.5) & (df_2023["evolucao_IDEP"] <= -0.2),
    df_2023["evolucao_IDEP"] < -0.5
]

categorias = [
    "Avanço significativo",
    "Avanço moderado",
    "Estabilidade",
    "Declínio moderado",
    "Declínio significativo"
]

df_2023["categoria_evolucao"] = np.select(condicoes, categorias, default="Sem classificação")
df_2023["crescimento_percentual"] = df_2024["crescimento_percentual"].round(3)

display(df_2023)
```

```python
#Quantide de escola por cada Distrito e diretoria de 2023

df_distrito=df_2023.groupby("Distrito").size()
df_diretoria =df_2023.groupby("Diretoria").size()

display(df_distrito)
display(df_diretoria)
```

```python
df_2023['crescimento_absoluto'] = df_2023['IDEP_FINAL'] - df_2023['IDEP_INICIAL']
df_2023['crescimento_percentual'] = (df_2023['crescimento_absoluto'] / df_2023['IDEP_INICIAL']) * 100
```

```python
#As top 10 Escolas Mais Avaliadas de 2023 
top5_escolas = (
    df_2023
    .sort_values("IDEP_FINAL", ascending=False)
    .head(10)
)

display(
    top5_escolas[
        ["Diretoria", "Escola", "IDEP_INICIAL", "IDEP_FINAL", "crescimento_percentual"]
    ].round(2)
)
```
